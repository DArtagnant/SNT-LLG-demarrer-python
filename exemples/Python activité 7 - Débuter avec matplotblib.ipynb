<h3> 2 avoir du style </h3><br/>
On peut changer le style utilisé par plot pour tracer ; <br/>
ce style est spécifié sous la forme d'une chaîne de caractères, <br/>par défaut 'b-', qui signifie une ligne bleue (b pour bleu, et - pour ligne). <br/>Ici on va préciser à la place ro, r qui signifie rouge et o qui signifie cercle

```python
plt.close() # pour effacer le  tracé précédent
plt.plot([1, 2, 3, 4, 5], [1, 4, 9, 16, 25], 'ro') # pour tracer les points rouge
plt.plot([1, 2, 3, 4, 5], [1, 4, 9, 16, 25], 'b-')  # pour tracer la courbe bleue
# ici on veut dire d'utiliser
# pour l'axe des X : entre 0 et 5
# pour l'axe des Y : entre -5 et 20
plt.axis([0, 5, -5, 20])  # format [xmin, xmax, ymin, ymax]
plt.show()
```

```python
plt.close()

# échantillon de points entre 0 et 5 espacés de 0.2
t = np.arange(0., 5., 0.2)
# plusieurs styles de ligne
plt.plot(t, t, 'r--')
plt.plot(t, t**2, 'bs')
plt.plot(t, t**3, 'g^')
# on pourrait ajouter d'autres plot bien sûr aussi
plt.show()
```

```python
plt.close()

# échantillon de points entre 0 et 5 espacés de 0.2
t = np.arange(0., 5., 0.2)
# la m^me chose en une ligne de code 
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
# on pourrait ajouter d'autres plot bien sûr aussi
plt.show()
```

<h3> Un peu d’explication </h3>
Np.arange  ( valeur initiale ; valeur finale , pas de calcul)<br/>

- Première courbe :  y =t   --> __r--  droite pointillé rouge__<br/>
- Deuxième courbe y = t²    --> __bs :  carré ( squarre) bleu__    <br/>
- Troisième courbe y= t^3   -->  __g^   cube triangle   ^  vert (g)__


<h3> 3 plusieurs courbes dans le même graphique  <b>  subplot() </b></h3>

```python
plt.close()
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)
## deux domaines presque identiques
# celui-ci pour les points rouges
t1 = np.arange(0.0, 5.0, 0.2)
# celui-ci pour la ligne bleue
t2 = np.arange(0.0, 5.0, 0.02)
# on crée un 'subplot'  courbe du haut
plt.subplot(2, 1, 1) # quadrillage de 2 lignes
#  1 colonne on travaille sur l' emplacement 1
# le fonctionnement de matplotlib est dit 'stateful'
# par défaut on dessine dans le dernier objet créé
plt.axis([0, 5, -1, 1]) # forçage du range 1
plt.plot(t1, f(t1), 'ro', t2, f(t2))
# une deuxième subplot, courbe du bas
plt.subplot(2,1,2)
# on écrit dedans
plt.axis([0, 6.28, -1.2 ,1.2]) # forçage du range 2
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()
```

<h3> 4 Courbe paramétrée   </b></h3>

```python
import matplotlib.pyplot as plt
import numpy as np
from math import exp
t = np.arange(0, 200, 0.1)  # plage de variation du paramètre
x = 10*np.cos(t)*np.exp(-0.01*t)  # équation de x(t)
y = 10*np.sin(t)*np.exp(-0.01*t) #  équation de y(t)
fig, ax = plt.subplots()         # déclaration de la courbe ax   
ax.plot(x, y)                   #  génération de la courbe paramétrée
plt.show()
```
